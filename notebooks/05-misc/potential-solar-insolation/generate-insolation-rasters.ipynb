```python
import os
import datetime as dt
import IPython

from PySAGA_cmd import (
    SAGA
)

import pandas as pd
import numpy as np

from joblib import Parallel, delayed

```

```python
# Instantiate SAGA session
saga_cmd = '/Applications/SAGA.app/Contents/MacOS/saga_cmd'
saga = SAGA(saga_cmd=saga_cmd)

# Choose a library
lighting = saga / 'ta_lighting'
pisr = lighting / 'Potential Incoming Solar Radiation'

# Print out tool options
print(pisr.execute(ignore_stderr=True).stdout)
```

```python
# Generate rasters 
latitude=34.003463

grd_total_out_dir = '/Users/patmccornack/Documents/ucsb_fog_project/data/gis/pisr_ts_rasters'
sample_grid = '/Users/patmccornack/Documents/ucsb_fog_project/data/gis/source_data/sm02_dem_clipped.tif'
times = np.arange(6, 20, 0.5).tolist()
datetimes = pd.date_range(start="2023-10-29",end="2023-11-30").tolist()
dates = [datetime.date() for datetime in datetimes]

for date in dates:
    for time in times:
        output = pisr.execute(
            GRD_DEM=sample_grid,
            GRD_TOTAL=os.path.join(grd_total_out_dir, f'rad_{str(date)}_{time}.tif'),
            LOCATION=1, # Calculate from grid system
            LATITUDE=latitude, # Probably unnecessary when LOCATION=1
            DAY=date,
            DAY_STOP=date,
            PERIOD=0,
            MOMENT=time,
            ignore_stderr=True
            )
```
